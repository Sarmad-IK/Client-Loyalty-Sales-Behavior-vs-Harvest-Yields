### **3.2 Finding and Handling Outliers**

**Dealing with myagro_seasons:**

`myagro_seasons` shows number of seasons that the client has farmed with myagro, which is an important column that can be associated with clients seniority level. To explore more on it, we need to detect the outliers.

```python
# distribution 
sns.histplot(x='myagro_seasons',data=harvest,kde=True,bins=40)
plt.title('Distribution of myagro_seasons',fontsize=14);
```

As you can see from the boxplot, the distribution is not normal therefore it is not approproaite to use standard deviation to detect outliers. Instead, we use interquartile range to remove outliers as it is the most suitable approach for skewed distribution.  

```python
# use 3*IQR to detect outliers
q3 = np.quantile(harvest['myagro_seasons'],0.75)
q1 = np.quantile(harvest['myagro_seasons'],0.25)
whis = (q3-q1)*3+q3
whis
```

We use (3*IQR + upper quartile) to identify extreme outliers and impossible values, then replacing them with median


```python
harvest['myagro_seasons'] = np.where(harvest['myagro_seasons'] >=whis, harvest['myagro_seasons'].median(), harvest['myagro_seasons'])
```

Since the distribution of `myagro_seasons` is skewed to right with the presense of outliers, which will disproportionately affect the value of mean. Therefore we use **median** to handling with the outliers.



### **3.3 Data Tranformation**

Yield is our main target. While yield values are collected by different crop types, each crop type have different measurement range of yield. To eliminate different scale effect on yield, we use 75th percentile to identify clients with high yield. Sepecifically, we create a column called 'high_yield_yes_no' to separate the clients with high yield or not based on 75th percentile for each crop type.

##### Creating `high_yield_yes_no` column based on 75th percentile

```python
# Creating yield grouped by crop descption table to extract 75th percentile yield values for each crop type
ycdf = harvest.groupby('crop')['yield'].describe()
ycdf
```

```python
# Storing 75th percentile values in appropiate varibles
arachide75 = ycdf['75%'][0]
mais75 =  ycdf['75%'][1]
mil75 =  ycdf['75%'][2]
riz75 =  ycdf['75%'][3]
sorgho75 =  ycdf['75%'][4]
```

```python
# Creating high_yield_yes_no column based on 75th percentile condition set for each crop type
harvest['high_yield_yes_no'] = np.where(((harvest['crop'] == 'arachide') &(harvest['yield'] >= arachide75)), 'yes', 'no')
harvest['high_yield_yes_no'] = np.where(((harvest['crop'] == 'mais') &(harvest['yield'] >= mais75)), 'yes', harvest['high_yield_yes_no'])
harvest['high_yield_yes_no'] = np.where(((harvest['crop'] == 'mais') &(harvest['yield'] >= mais75)), 'yes', harvest['high_yield_yes_no'])
harvest['high_yield_yes_no'] = np.where(((harvest['crop'] == 'mil') &(harvest['yield'] >= mil75)), 'yes', harvest['high_yield_yes_no'])
harvest['high_yield_yes_no'] = np.where(((harvest['crop'] == 'riz') &(harvest['yield'] >= riz75)), 'yes', harvest['high_yield_yes_no'])
harvest['high_yield_yes_no'] = np.where(((harvest['crop'] == 'sorgho') &(harvest['yield'] >= sorgho75)), 'yes', harvest['high_yield_yes_no'])
```
